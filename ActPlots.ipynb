# Actividad Mapas de calor y boxplots
A continuacion se presenta un código para analizar datos dentro de un conjunto de valores que muestran los resultados de un análisis de riesgo crediticio.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Importar archivo csv, visualizar información general. **

```python
path = "credit_risk.csv"
df = pd.read_csv(path)
print(df.describe())
```

**Gráfico mapa de calor para visualización de la correlación entre variables**

```python
#sns.color_palette("mako", as_cmap=True)
df_corr = df.corr()
sns.heatmap(data=df_corr, cmap="mako")
```

```python
sns.boxplot(df, y = "Age", color="orange")
```

```python
sns.boxplot(df, y = "Amount", color="green")
```

```python
sns.boxplot(df, y = "Rate")
```

```python
sns.boxplot(df, y = "Income", color="red")
```

```python
sns.displot(df, x = "Rate", kde=True)
```

```python
sns.catplot(data=df, x="Income", y="Amount", hue="Status")
```

```python
sns.regplot(x="Amount", y="Rate", data=df)
```

```python
sns.displot(df, x= "Age", stat="probability")
```
